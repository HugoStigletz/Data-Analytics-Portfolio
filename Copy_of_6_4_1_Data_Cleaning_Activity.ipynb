# Step 4: Create a Pandas DataFrame from the CSV file

````python
customers = pd.read_csv('6.4.1 customers.csv')

customers.head()
````

# Step 5: How many missing values are in the data set?

It is common for data sets to have missing values. Since this is a small data set, we can see the places where data is missing, but there is a function `.isnull()` that helps us find missing data.

````python
customers.isnull()
````

# Step 6: Fill NaN values
Use `fillna()` to fill in the values.

Assume `Tina` is 37 years old and the missing name is `Jack`.

````python
#Fill in Name column
customers["Name"].fillna("Jack", inplace = True)

#Fill in Age column
customers["Age"].fillna(37, inplace = True)

#Fill in Gender column
customers["Gender"].fillna("f", inplace = True)
````

````python
customers.head(10)
````

# Step 7: What are the data types?
* Sometimes we need our data in a certain data type to work with. Use .dtypes to see what types are in the `customers` data.

````python
#Step 7

customers.dtypes
````

# Step 8: Change `Age` column from a float to int


````python
#Step 8

customers['Age'] = customers['Age'].astype('int64')
````

# Step 9: Drop duplicates
* We can see from the code cell below there are duplicate rows. Drop the duplicates in the following cell.

````python
customers.duplicated().sum()
````

````python
#Step 9
# customers = customers.drop_duplicates() OR
customers.drop_duplicates(inplace= True)
customers
````

# Step 10: `replace()` demo
* We will go more into string manipulation in the next module. Run the code cell below to change `Jane@#!!!!!!!` to `Jane`.

````python
customers['Name'] = customers['Name'].str.replace('[!@#]','')
customers
````

````python
#Step 10 A longer version of the code above
import warnings
warnings.filterwarnings('ignore')

customers['Name'] = customers['Name'].str.replace('!','').str.replace('@','').str.replace('#','')
customers
````
